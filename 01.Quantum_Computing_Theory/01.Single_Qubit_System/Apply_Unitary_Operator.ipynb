```python
U = np.array([[1, 1], [1, -1]]) / np.sqrt(2)

initial_state = np.array([3,2])
#normalizing the state:
normalized_state = initial_state/np.linalg.norm(initial_state)

final_state = apply_unitary(U, initial_state)

print(f'Initial state: {normalized_state}')
print(f'Final state: {final_state}')

#normalizing the state:
normalized_final_state = final_state #/np.linalg.norm(final_state)
```
